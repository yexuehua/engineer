```python
import plotly.express as px
from dash import Dash, dcc, html
from dash.dependencies import Input, Output
import pandas as pd

# Load the Titanic dataset
df = test_df

# Initialize the Dash app
app = Dash(__name__)

# Layout of the app
app.layout = html.Div([
    dcc.Dropdown(
        id='x-axis-dropdown',
        options=[{'label': col, 'value': col} for col in df.columns if col not in ['Name', 'Ticket', 'Cabin']],
        value='Age'
    ),
    dcc.Dropdown(
        id='y-axis-dropdown',
        options=[{'label': col, 'value': col} for col in df.columns if col not in ['Name', 'Ticket', 'Cabin']],
        value='Fare'
    ),
    dcc.Graph(id='scatter-plot')
])

# Callback to update the scatter plot based on dropdown selection
@app.callback(
    Output('scatter-plot', 'figure'),
    Input('x-axis-dropdown', 'value'),
    Input('y-axis-dropdown', 'value')
)
def update_scatter_plot(x_axis, y_axis):
    fig = px.scatter(df, x=x_axis, y=y_axis, title=f'Scatter plot of {x_axis} vs {y_axis}')
    return fig

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
import plotly.express as px
import plotly.graph_objects as go
from dash import Dash, dcc, html
from dash.dependencies import Input, Output
import pandas as pd
import numpy as np

df = test_df
# Initialize the Dash app
app = Dash(__name__)

# Layout of the app
app.layout = html.Div([
    dcc.Graph(
        id='heatmap',
        figure=go.Figure(
            data=go.Heatmap(
                z=df.values,
                x=df.columns,
                y=df.index,
                colorscale='Viridis'
            ),
            layout=go.Layout(
                title='Interactive Heatmap',
                xaxis_title='Columns',
                yaxis_title='Rows'
            )
        )
    ),
    html.Div(id='output-div')
])

# Callback to update the output div based on heatmap click
@app.callback(
    Output('output-div', 'children'),
    Input('heatmap', 'clickData')
)
def display_click_data(clickData):
    if clickData is None:
        return 'Click on a cell in the heatmap to see the details here.'
    else:
        point = clickData['points'][0]
        x_value = point['x']
        y_value = point['y']
        z_value = point['z']
        return f'You clicked on cell with value: {z_value}, at column: {x_value}, row: {y_value}.'

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```
